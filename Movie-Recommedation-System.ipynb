```python
# Importing Libraries
import numpy as np
import pandas as pd
```

```python
# Importing Data Sets
movies = pd.read_csv('Movies Data.csv')
credits = pd.read_csv('Credits Data.csv') 
```

```python
movies = movies.merge(credits,on='title')
```

```python
# Selecting Required Columns Only
movies.head()
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
# Changing Data Set
movies = movies[['Movie_id_x','title','overview','Genres','Keywords','Cast','Crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
# Selecting Specific Words From The Columns
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# Removing NULL Values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# Selecting Genres
movies['Genres'] = movies['Genres'].apply(convert)
movies.head()
```

```python
# Selecting Keywords
movies['Keywords'] = movies['Keywords'].apply(convert)
movies.head()
```

```python
# Selecting First 3 Names From Cast
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['Cast'] = movies['Cast'].apply(convert)
movies.head()
```

```python
movies['Cast'] = movies['Cast'].apply(lambda x:x[0:3])
```

```python
# Selecting Only Directors Name From Crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['Crew'] = movies['Crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
# Removing Spaces Between Names To Avoid Vector Conflict
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['Cast'] = movies['Cast'].apply(collapse)
movies['Crew'] = movies['Crew'].apply(collapse)
movies['Genres'] = movies['Genres'].apply(collapse)
movies['Keywords'] = movies['Keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Merging Various Columns
# Making a column name "tags" that contain 'overview','Genres','Keywords','Casr','Crew' 
movies['tags'] = movies['overview'] + movies['Genres'] + movies['Keywords'] + movies['Cast'] + movies['Crew']
```

```python
new = movies.drop(columns=['overview','Genres','Keywords','Cast','Crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
# Now We Vectorize Our Data With Features
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
# Calculating similarities on the basis of cosine angle between the vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
# Movie Index
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
# Recommend Closest 5 Movies Based On Similarity
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
#Testing
recommend('Spider-Man')
```
